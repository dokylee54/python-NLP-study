# W에 대한 cost 함수 시각화하기

```python
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
```

```python
X = [1, 2, 3]
Y = [1, 2, 3]
```

```python
W = tf.placeholder(tf.float32)
```

```python
hypothesis = X * W
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
W_val = []
cost_val = []
```

```python
for i in range(-30, 50):
    feed_W = i * 0.1
    curr_cost, curr_W = sess.run([cost, W], feed_dict={W: feed_W})
    W_val.append(curr_W)
    cost_val.append(curr_cost)
```

```python
plt.plot(W_val, cost_val)
plt.show()
```

> 위의 그래프의 각 점에서의 기울기가<br>
'+' : W를 - 방향으로 움직임<br>
'-' : W를 + 방향으로 움직임

# Minimize cost

```python
x_data = [1, 2, 3]
y_data = [1, 2, 3]
```

```python
W = tf.Variable(tf.random_normal([1]), name='weight')
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
```

```python
hypothesis = X * W
```

```python
cost = tf.reduce_sum(tf.square(hypothesis - Y))
```

```python
# Minimize: Gradient Descent using derivative: W -= learning_rate * derivative
learning_rate = 0.1
gradient = tf.reduce_mean(( W * X - Y ) * X)
descent = W - learning_rate * gradient
update = W.assign(descent)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(21):
    sess.run(update, feed_dict={X: x_data, Y: y_data})
    print(step, sess.run(cost, feed_dict={X: x_data, Y: y_data}), sess.run(W))
```

### 위에선 cost 함수를 직접 미분해서 gradient를 구했지만, optimizer 사용시 안그래도 됨

```python
X = [1, 2, 3]
Y = [1, 2, 3]
```

```python
W = tf.Variable(5.0)    # W에 5.0을 주고 시작
```

```python
hypothesis = X * W
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(100):
    print(step, sess.run(W))
    sess.run(train)
```

### 만약 optimizer 내부에서 계산되는 descent 값을 내가 조정하고 싶으면

```python
X = [1, 2, 3]
Y = [1, 2, 3]
```

```python
W = tf.Variable(5.0)
```

```python
hypothesis = X * W
```

```python
gradient = tf.reduce_mean(( W * X - Y ) * X) * 2
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
```

```python
# Get gradients
gvs = optimizer.compute_gradients(cost)
```

```python
# Optional: modify gradient if necessary
# gvs = [(tf.clip_by_value(grad, -1., 1.), var) for grad, var in gvs]
```

```python
# Apply gradients
apply_gradients = optimizer.apply_gradients(gvs)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(100):
    print(step, sess.run([gradient, W]))  # [gradient, W, gvs]처럼 gvs을 출력하려고 하면 오류
    sess.run(apply_gradients)
```
